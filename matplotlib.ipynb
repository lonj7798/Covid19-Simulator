## Pyplot Tutorial

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4, 5, 6, 7])
plt.ylabel('numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 9, 20])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
# plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.plot(t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.plot(t, t**2, 'r--',
         t, t**3, 'bs',
         t, t, 'g^')
plt.show()
```

## Category

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131) # length / height / position
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(233) # length / height / position
plt.bar(names, values)

plt.subplot(232)
plt.scatter(names, values)

plt.subplot(231)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211) # if length is larger than height, the position of graphs are row
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(211) # same position -> overlaps
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure()
plt.subplot(122)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(121) # if height is larger than length, grahps position is col
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
